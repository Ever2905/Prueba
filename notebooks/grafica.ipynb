```python
import pandas as pd
import matplotlib.pyplot as plt

# Definir los datos
tiempo = [0, 10, 20, 30, 40, 50]
posicion = [30, 52, 38, 0, -37, -53]

# Crear un DataFrame con los datos
data = {'Tiempo': tiempo, 'Posición': posicion}
df = pd.DataFrame(data)

# Visualizar el DataFrame
print(df)

# Plotear la tabla posición vs tiempo


plt.plot(df['Tiempo'], df['Posición'], marker='o', linestyle='-', color='red')

plt.xlabel('Tiempo')

plt.ylabel('Posición')
plt.title('Posición vs Tiempo')
plt.grid(True)
plt.show()
```
